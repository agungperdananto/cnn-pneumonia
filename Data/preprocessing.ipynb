```python
import os, cv2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
PATH = os.getcwd()
# Define data path
data_path = PATH + '/test'
data_dir_list = os.listdir(data_path)
```

```python
#inisialisasi
img_row = 28
img_col = 28
num_channel = 1
img_data_list= []

##Mencari dan mengubah masing-masing gambar dalam tiap2 folder, 
##lalu mengubah sesuai format dan ukuran standard yang ditentukan 
for dataset in data_dir_list:
    img_list=os.listdir(data_path+'/'+ dataset) #Mencari dan mendeteksi tiap-tiap folder
    print ('Loaded the images of dataset-'+'{}\n'.format(dataset))
    for img in img_list:
        input_img=cv2.imread(data_path + '/'+ dataset + '/'+ img ) #membaca masing-masing data gambar dan mengubah ke dalam biner
        input_img=cv2.cvtColor(input_img, cv2.COLOR_BGR2GRAY) #mengubah tiap-tiap gambar menjadi format warna grayscale
        input_img_resize=cv2.resize(input_img,(img_row,img_col)) #resize ukuran gambar sesuai dengan img_row dan img_col
        input_img_resize = input_img_resize.reshape(img_row*img_col) #mengubah ukuran matriks menjadi satu baris contoh: dari 50x50 menjadi 1x2500
        img_data_list.append(input_img_resize) #menambahkan baris diatas ke dalam data list(img_data_list)

img_data = np.array(img_data_list) #mengubah img_data_list yang berbentuk list menjadi bentuk array
img_data = img_data.astype('float32') #mengubah setiap elemen data img_data menjadi float32
print (img_data.shape)
```

```python
#img data dalam bentuk dataframe
df0 = pd.DataFrame(img_data)
df0
```

```python
labels=[]
#Menentukan label pada data sesuai dengan jumlah data pada masing-masing folder
#satu folder merepresentasikan label tertentu
lbl1 = np.ones(390)
lbl2 = np.zeros(234)
labels.append(lbl1)
labels = np.array(labels)
labels.resize()
for i in range(len(lbl2)):
    labels = np.append(labels, lbl2[i])
#labels = labels.reshape(20)

labels = labels.reshape(624,1)

print(labels.shape)
labels
```

```python
#menambahkan label yang sudah dibuat ke dalam data
img_data_label = np.insert(img_data, [0], labels, axis=1)
img_data_label
```

```python
#shuffling baris (axis x)
np.random.shuffle(img_data_label)
img_data_label
```

```python
# menampilkan dalam bentuk dataframe data yang sudah diberi label
df = pd.DataFrame(img_data_label)
df.head()
```

```python
#menambahkan header pada tiap-tiap kolom
col_header = ['label']
for i in range (784):
    pixel = 'pixel ' +str(i)
    col_header.insert((i+1),pixel)
col_header
#img_data_label.shape
df = pd.DataFrame(img_data_label, columns=col_header)
df.head()
```

```python
#konvert ke dalam format csv
df.to_csv('data_train_mnist.csv', sep=',', encoding='utf-8', index=False)
```
